```python
import pandas as pd

nanchang_weather = pd.read_csv('nanchang.csv', sep=',', encoding='GB2312')
beijing_weather = pd.read_csv('beijing.csv', sep=',', encoding='GB2312')
hangzhou_weather = pd.read_csv('hangzhou.csv', sep=',', encoding='GB2312')
wuhan_weather = pd.read_csv('wuhan.csv', sep=',', encoding='GB2312')
nanchang_weather['日期'] = pd.to_datetime(nanchang_weather['日期'],format="%Y/%m/%d")  #将字符串改为时间格式
beijing_weather['日期'] = pd.to_datetime(beijing_weather['日期'],format="%Y/%m/%dT")  #将字符串改为时间格式
hangzhou_weather['日期'] = pd.to_datetime(hangzhou_weather['日期'],format="%Y/%m/%dT")  #将字符串改为时间格式
wuhan_weather['日期'] = pd.to_datetime(wuhan_weather['日期'],format="%Y/%m/%dT")  #将字符串改为时间格式
nanchang_weather = nanchang_weather.set_index('日期') # 将date设置为index
beijing_weather = beijing_weather.set_index('日期') # 将date设置为index
hangzhou_weather = hangzhou_weather.set_index('日期') # 将date设置为index
wuhan_weather = wuhan_weather.set_index('日期') # 将date设置为index
```

```python
nanchang_weather17 = nanchang_weather['2017'] 
beijing_weather17 = beijing_weather['2017'] 
hangzhou_weather17 = hangzhou_weather['2017'] 
wuhan_weather17 = wuhan_weather['2017'] 

nanchangAQI_min17 = nanchang_weather17['AQI指数'].min()
nanchangAQI_max17 = nanchang_weather17['AQI指数'].max()
nanchangAQI_mean17 = nanchang_weather17['AQI指数'].mean()

nanchang_weather18 = nanchang_weather['2018'] 
beijing_weather18 = beijing_weather['2018'] 
hangzhou_weather18 = hangzhou_weather['2018'] 
wuhan_weather18 = wuhan_weather['2018'] 

nanchangAQI_min18 = nanchang_weather18['AQI指数'].min()
nanchangAQI_max18 = nanchang_weather18['AQI指数'].max()
nanchangAQI_mean18 = nanchang_weather18['AQI指数'].mean()

beijingAQI_min18 = beijing_weather18['AQI指数'].min()
beijingAQI_max18 = beijing_weather18['AQI指数'].max()
beijingAQI_mean18 = beijing_weather18['AQI指数'].mean()

hangzhouAQI_min18 = hangzhou_weather18['AQI指数'].min()
hangzhouAQI_max18 = hangzhou_weather18['AQI指数'].max()
hangzhouAQI_mean18 = hangzhou_weather18['AQI指数'].mean()

wuhanAQI_min18 = wuhan_weather18['AQI指数'].min()
wuhanAQI_max18 = wuhan_weather18['AQI指数'].max()
wuhanAQI_mean18 = wuhan_weather18['AQI指数'].mean()

nanchang_weather17['AQI指数'].describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] #SimHei是黑体的意思
#fig, axes = plt.subplots(figsize = (10, 10)) #4个没有数据的图表
data = [{'min':nanchangAQI_min18,'mean':nanchangAQI_mean18,'max':nanchangAQI_max18},
        {'min':beijingAQI_min18,'mean':beijingAQI_mean18,'max':beijingAQI_max18},
        {'min':hangzhouAQI_min18,'mean':hangzhouAQI_mean18,'max':hangzhouAQI_max18},
        {'min':wuhanAQI_min18,'mean':wuhanAQI_mean18,'max':wuhanAQI_max18}]
df = pd.DataFrame(data, index=['nanchang', 'beijing', 'hangzhou', 'wuhan'])
#s.plot(kind='bar',color = 'k',grid = True,alpha = 0.5,ax = axes[0])  # ax参数 → 选择第几个子图  # 单系列柱状图方法一：plt.plot(kind='bar/barh')
print(df)
```
