```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] #SimHei是黑体的意思
weather = pd.read_csv('nanchang.csv', sep=',' ,encoding='GB2312')
weather#.head()
weather['日期'] =pd.to_datetime(weather['日期'],format="%Y/%m/%dT")

weather = weather.set_index('日期') # 将date设置为index
weather18 = weather['2018']    #按日期筛选数据
weather17 = weather['2017'] 
#print(df[‘2016‘:‘2017‘].head(2)) #获取2016至2017年的数据 
#print(df.to_period(‘Q‘).head()) #按季度显示，但不统计

diff18 = weather18.groupby('质量等级').count()
diff17 = weather17.groupby(['质量等级']).count()#.reset_index(name='days')

mild_pollution=weather17[weather17['质量等级']=='轻度污染']['质量等级'].count()
middle_pollution=weather17[weather17['质量等级']=='中度污染']['质量等级'].count()
heavy_pollution=weather17[weather17['质量等级']=='重度污染']['质量等级'].count()
fine=weather17[weather17['质量等级']=='良']['质量等级'].count()
good=weather17[weather17['质量等级']=='优']['质量等级'].count()
#diff17
#weather18
labels='轻度污染','中度污染','重度污染','优','良'

sizes=mild_pollution,middle_pollution,heavy_pollution,fine,good

colors='#FFEFD5','#FFDEAD','#FFA54F','#CAE1FF','#B0E0E6'

explode=0,0,0.1,0.1,0

plt.pie(sizes,explode=explode,labels=labels,

        colors=colors,autopct='%1.1f%%',shadow=True,startangle=0)

plt.axis('equal')
plt.title('南昌市2017年空气质量状况饼图')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] #SimHei是黑体的意思

nanchang_weather = pd.read_csv('nanchang.csv', sep=',', encoding='GB2312')
beijing_weather = pd.read_csv('beijing.csv', sep=',', encoding='GB2312')
hangzhou_weather = pd.read_csv('hangzhou.csv', sep=',', encoding='GB2312')
wuhan_weather = pd.read_csv('wuhan.csv', sep=',', encoding='GB2312')
nanchang_weather['日期'] = pd.to_datetime(nanchang_weather['日期'],format="%Y/%m/%d")  #将字符串改为时间格式
beijing_weather['日期'] = pd.to_datetime(beijing_weather['日期'],format="%Y/%m/%dT")  #将字符串改为时间格式
hangzhou_weather['日期'] = pd.to_datetime(hangzhou_weather['日期'],format="%Y/%m/%dT")  #将字符串改为时间格式
wuhan_weather['日期'] = pd.to_datetime(wuhan_weather['日期'],format="%Y/%m/%dT")  #将字符串改为时间格式
nanchang_weather = nanchang_weather.set_index('日期') # 将date设置为index
beijing_weather = beijing_weather.set_index('日期') # 将date设置为index
hangzhou_weather = hangzhou_weather.set_index('日期') # 将date设置为index
wuhan_weather = wuhan_weather.set_index('日期') # 将date设置为index
nanchang_weather18 = nanchang_weather['2018'] 
beijing_weather18 = beijing_weather['2018'] 
hangzhou_weather18 = hangzhou_weather['2018'] 
wuhan_weather18 = wuhan_weather['2018'] 


nan_diff18 = nanchang_weather18.groupby('质量等级').count()
bei_diff18 = beijing_weather18.groupby('质量等级').count()
hang_diff18 = hangzhou_weather18.groupby('质量等级').count()
wu_diff18 = wuhan_weather18.groupby('质量等级').count()

```

```python
nan_qua = nan_diff18["AQI指数"]
bei_qua = bei_diff18["AQI指数"]
hang_qua = hang_diff18["AQI指数"]
wu_qua = wu_diff18["AQI指数"]
```

```python
d = {'南昌' : pd.Series(nan_qua),
     '北京' : pd.Series(bei_qua),
     '杭州' : pd.Series(hang_qua),
     '武汉' : pd.Series(wu_qua),}

df = pd.DataFrame(d)
```

```python
df.plot.barh(alpha=0.7,legend=True,colormap='Set2') 
plt.legend(loc=1)
plt.grid(True)
plt.xlabel("天数")
plt.title('各城市空气质量状况天气对比',fontsize='15')
```
